```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://wikisource.org/wiki/Main_Page"
```

```python
response = requests.get(url)
```

```python
response
```

```python
data = response.text
```

```python
data
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
soup
```

```python
tags = soup.find_all('a')
```

```python
tags
```

```python
for tag in tags:
    print(tag.get('href'))
```
